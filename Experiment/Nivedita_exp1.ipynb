```python
X = df.drop(columns='is_play', axis=1)
Y = df['is_play']
```

```python
x = X.to_numpy()
y = Y.to_numpy()
```

```python
def hypothesis(x,y):
  for i, val in enumerate(y):
    if val == 'yes':
      specific_h = x[i].copy()

  for i, val in enumerate(x):
    if y[i] == 'yes':
      for j in range(len(specific_h)):
        if val[j] != specific_h[j]:
          specific_h[j] = '?'
      else:
        pass
  return specific_h


hypothesis(x,y)

```

```python
def generalize_hypothesis(x,y):
  specific_h = hypothesis(x,y)
  general_h = [['?' for i in range(len(specific_h))] for i in range(len(specific_h))]
generalize_hypothesis
```
